```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import polars as pl
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import Matern
```

```python
start = -1000
n = 100

data = pd.read_csv("GBPUSD_M1.csv")

priceData = pl.DataFrame(data[["Time", "Close"]]).select([
    pl.col("Time").str.strptime(pl.Datetime, "%Y-%m-%d %H:%M"),
    pl.col("Close")
])

priceData = (
    priceData.sort(by="Time")
    .group_by_dynamic("Time", every="1h")
    .agg(pl.col("Close").median())
    .drop_nulls()
)

resultIndices = []
predictedMeans = []
predictedStds = []

for i in range(start-n,start):
    cutoff = i
    project = 2
    
    rangeData = priceData[-n+cutoff:cutoff]
    
    X = rangeData.select(
        (pl.col("Time").dt.timestamp() / (1000000 * 60 * 60)) 
    ).to_numpy()
    y = rangeData["Close"].to_numpy()
    
    
    X = [val - X[-1] for val in X]
    
    yMean = y.mean()
    kernel = 1.0 * Matern(length_scale=1.0, nu=1.5)
    gpr = GaussianProcessRegressor(kernel=kernel, normalize_y=False, n_restarts_optimizer=0)
    gpr.fit(X, y-yMean)
    
    Xtest = np.linspace(np.min(X),project,100).reshape(-1,1)
    meanPred, stdPred = gpr.predict(Xtest, return_std=True)
    meanPred += yMean
    
    resultIndices.append(cutoff)
    predictedMeans.append(meanPred[-1])
    predictedStds.append(stdPred[-1])


resultIndices = np.array(resultIndices)
predictedMeans = np.array(predictedMeans)
predictedStds = np.array(predictedStds)

#plt.plot(X,y,color='teal')
#plt.plot(Xtest, meanPred, color='darkorange')

plt.fill_between(
    resultIndices,
    predictedMeans - 1.96 * predictedStds,
    predictedMeans + 1.96 * predictedStds,
    alpha=0.5,
    color='darkorange'
)

cutoff = start
project = 0

rangeData = priceData[-n+cutoff:cutoff]

X = rangeData.select(
    (pl.col("Time").dt.timestamp() / (1000000 * 60 * 60)) 
).to_numpy()
y = rangeData["Close"].to_numpy()

X = [val - X[-1] for val in X]

yMean = y.mean()
sampleNoise = 1e-3
kernel = 1.0 * Matern(length_scale=1.0, nu=1.5)
gpr = GaussianProcessRegressor(kernel=kernel, alpha=sampleNoise**2, normalize_y=False, n_restarts_optimizer=2)
gpr.fit(X, y-yMean)

Xtest = np.linspace(np.min(X),project,100).reshape(-1,1)
meanPred, stdPred = gpr.predict(Xtest, return_std=True)
meanPred += yMean

X = [val+cutoff for val in X]

plt.plot(X, y, color="teal")

print(gpr.kernel_)
for hp in kernel.hyperparameters:
    print(hp,"\n")
```
